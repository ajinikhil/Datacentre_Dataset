# Goal: Extract data centre information from peeringDB

```python
import requests
import json
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
api_key = os.getenv("PEERING_DB_API")
```

```python
headers = {
    "Authorization": f"Api-Key {api_key}"
}
```

```python
response = requests.get("https://www.peeringdb.com/api/fac", headers=headers)
```

```python
data = response.json()
```

```python
data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data["data"])
```

```python
print(f"Missing values in campus_id: {df['campus_id'].isna().mean() * 100}%")
print(f"Missing values in property: {df['property'].isna().mean() * 100}%")
```

## Data cleaning

```python
df = df.drop(columns= ["campus_id", "aka", "name_long", "social_media",  "clli",
                        "rencode", "npanxx", "notes", "sales_email", "sales_phone", 
                        "tech_email", "tech_phone", "available_voltage_services",
                        "diverse_serving_substations", "status_dashboard", "logo",
                        "address2", "floor", "suite"])
```

```python
df.to_csv("../datasets/peeringDB/datacentre_list.csv")
```
